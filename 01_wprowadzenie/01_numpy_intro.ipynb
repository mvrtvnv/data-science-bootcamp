maski logiczne

```python
A = np.arange(start=-10, stop=10, step=0.5)
A = A.reshape(10,-1)
A
```

maski logiczne służą np do identyfikacji wartości ujmenych i wycięcia ich z tablicy

```python
A >0
```

```python
A[A>0]
```

możemy łączyć ze sobą dwa warunki np. wartości większe id -5 i mniejsze od 5, nie możemy tego jednak zrobić w formie A[A>-5 and A<5] numpy zwróci nam błąd, poprawna metoda poniżej:

```python
np.bitwise_and(A > -5, A < 5)
```

funkcja zwróci nam maskę logiczną, prawda w miejscu gdy oba warunki są spęlnione, dzięku tej masce jesteśmy w stanie wyciąć elementy z przedziały -5 do 5:

```python
A[np.bitwise_and(A > -5, A < 5)]
```

alternatywą dla bitwise_and będzie bitwise_or -> warunek prawda jeżeli wartości są mniejsze od -5 lub większe od 5:

```python
np.bitwise_or(A < -5, A > 5)
```

```python
A[np.bitwise_or(A < -5, A > 5)]
```
